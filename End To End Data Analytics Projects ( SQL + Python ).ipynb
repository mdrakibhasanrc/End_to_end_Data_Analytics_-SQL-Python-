## Data Cleaning & Preprocessig

```python
# check duplicated values
df.duplicated().sum()
```

```python
# to check null values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head(2)
```

```python
df.rename(columns={'Order Id':'order_id','Order Date':'order_date','Postal Code':'postal_code','Sub Category':'sub_category'},inplace=True)
```

```python
df.rename(columns={'Product Id':'product_id','cost price':'cost_price','List Price':'list_price','Discount Percent':'discount_price'},inplace=True)
```

```python
df.columns=df.columns.str.lower()
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df['discount']=df['list_price']*df['discount_price']*0.01
```

```python
df['sale']=df['list_price']-df['discount']
```

```python
df['profit']=df['sale']-df['cost_price']
```

```python
df['month_name']=df['order_date'].dt.month_name()
```

```python
df['day_name']=df['order_date'].dt.day_name()
```

```python
df['quarter']=df['order_date'].dt.quarter
```

```python
df.head(2)
```

```python
df.drop(columns=['cost_price','list_price','discount_price'],inplace=True)
```

## Data Export to PostgressSQL

```python
from sqlalchemy import create_engine
```

```python
connection= 'postgresql://postgres:1234@localhost:5432/SQL_Practice'
```

```python
engine = create_engine(connection)
```

```python
df.to_sql('sales_data', con=engine, if_exists='replace', index=False)
```
